```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
import time
from sklearn.svm import SVC
from sklearn.multiclass import OneVsOneClassifier
```

```python
# Immport data
X = np.loadtxt('./archive/dataX.csv', delimiter=",")
y = np.genfromtxt('./archive/datay.csv', dtype="str", encoding="utf-8")
```

```python
chars, counts = np.unique(y, return_counts=True)
# Map chars to ints in interval [0,number_of_lables] 
mapping_dict = dict(zip(chars, [x for x in range(chars.size)]))
mapping_dict_backwards = dict(zip([x for x in range(chars.size)], chars)) #may be needed later

y = np.vectorize(mapping_dict.get)(y).astype("uint8")
X = X/255
```

Lets test a variety of hyper paramaters.

```python
num_tests = 1
kernals = ['rbf','linear', 'poly', 'sigmoid', 'precomputed']
c_vals = [.02, 1.3, 2, 5] #first batch [.05, .1, .7, 1] lead to wanting to try more, will later concat data frames to analyze full batch
results_df = pd.DataFrame(index=range(num_tests*len(kernals)), columns=["kernal", "reg_param", "validation_accuracy", "training_time", "model_object"])

i = 0
errors = 0
print("models completed:")
for test in range(num_tests):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

    for kern in kernals:
        for c in c_vals:
            try:
                ava_model = OneVsOneClassifier(SVC(C=c, kernel=kern))

                start = time.time()
                ava_model.fit(X_train, y_train)
                end = time.time()

                results_df.loc[i] = [kern, c, np.sum(ava_model.predict(X_test) == y_test)/y_test.shape[0], end-start, ava_model]
                i += 1
        
                print(str(i+1)+'/'+str(num_tests * len(kernals) * len(c_vals)), end=" ")
            except:
                errors += 1
```

Training resulted in 4 errors, always use try/axcept :)
